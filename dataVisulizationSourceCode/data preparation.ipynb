```python
import pandas as pd

# Load the Excel file
file_path = "scams.xlsx"  # Replace with your file path
df = pd.read_excel(file_path)

# Split the 'Date' column into 'Year', 'Month', 'Day'
df[['Year', 'Month', 'Day']] = df['Date'].astype(str).str.split('-', expand=True)

# Drop the original 'Date' column
df = df.drop(columns=['Date'])

# Aggregate the number of reports by 'Year' and 'Month'
# Assuming there's a 'ReportCount' or similar column indicating the number of reports
# If not, you can simply count the occurrences by using .size()
agg_df = df.groupby(['Year', 'Month']).size().reset_index(name='ReportCount')

# Save the aggregated data to a new Excel file
agg_df.to_csv("aggregated_reports_by_year_month.xlsx", index=False)

# Display the aggregated dataframe
print(agg_df.head())
```

```python
import pandas as pd

# Load the Excel file
file_path = "scams.xlsx"  # Replace with your actual file path
df = pd.read_excel(file_path)

# Assuming the 'Date' and 'State' columns are present
# Split the 'Date' column into 'Year', 'Month', and 'Day'
df[['Year', 'Month', 'Day']] = df['Date'].astype(str).str.split('-', expand=True)

# Aggregate the data by 'Year', 'Month', and 'State'
agg_df = df.groupby(['Year', 'Month', 'State']).size().reset_index(name='ReportCount')

# Save the aggregated data to a new Excel file
csv_file_path = "aggregated_reports_by_year_month_state.csv"
agg_df.to_csv(csv_file_path, index=False)

# Display the first few rows of the aggregated data
print(agg_df.head())
```

```python
import pandas as pd

# Load the Excel file
file_path = "scams.xlsx"  # Replace with your file path
df = pd.read_excel(file_path)

# Aggregate the data to get the total number of reports per state
total_reports_by_state = df.groupby('State')['Number of Reports'].sum().reset_index()

# Aggregate the data to get the number of reports per scam type for each state
scam_reports_by_state = df.groupby(['State', 'Scam Type'])['Number of Reports'].sum().unstack(fill_value=0)

# Merge the total reports with the scam type breakdown
tooltip_data = total_reports_by_state.set_index('State').join(scam_reports_by_state)

# Display the aggregated data
print(tooltip_data)

# Optionally, save the aggregated data to a CSV file for use in D3.js
tooltip_data.to_csv("aggregated_reports_for_tooltip.csv")
```

```python
import pandas as pd

# Load the Excel file
file_path = "scams.xlsx"  # Update the path
df = pd.read_excel(file_path)

# Aggregate data by scam type and gender
gender_data = df.groupby(['Scam Type', 'Gender'])['Number of Reports'].sum().unstack().fillna(0)

# Save the processed data to a CSV file
gender_data.to_csv("scam_gender_data.csv")
```
